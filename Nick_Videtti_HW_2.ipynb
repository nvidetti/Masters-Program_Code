This assignment will take two documents and perform a sentence tokenization on them, then will calculate some descriptive statistics corresponding to those sentences.

More will follow once the documents have been properly tokenized, but let's focus on that first!
<br>
There are three fundamental steps to this initial tokenizaion:
<br>
&emsp;1. Tokenizing the raw text. This will help split the documents into a list of its sentences.
<br>
&emsp;2. Tag the sentences with Part of Speech (POS) tags.
<br>
&emsp;3. Filter each document to only include sentences that contain adjectives and/or adverbs. These will correspond with POS tags of "JJ", "JJR", "JJS" for adjectives and "RB", "RBR", "RBS", "WRB" for adverbs.

```python
#Import the Natural Language Toolkit and the nltk.book package
import nltk
from nltk import book

#Monty Python and the Holy Grail sentence tokens
#Turn word tokens into one string of raw text
Monty_Python_text = ' '.join([word for word in nltk.book.text6.tokens])
#Tokenize sentences
Monty_Python_sentences = nltk.sent_tokenize(Monty_Python_text)

#Knicks History with tokenizing
#source: Wikipedia - https://en.wikipedia.org/wiki/History_of_the_New_York_Knicks
with open('C:/Users/nvidetti/Downloads/History of Knicks Wiki.txt','r', encoding = 'utf-8') as Knicks_History_Text_File: Knicks_History = Knicks_History_Text_File.read()
Knicks_History_sentences = nltk.sent_tokenize(Knicks_History)
```

```python
#Tokenze the words in each sentence for Monty Python sentences
Monty_Python_sentences = [nltk.word_tokenize(sent) for sent in Monty_Python_sentences]

#Tokenze the words in each sentence for Knicks History sentences
Knicks_History_sentences = [nltk.word_tokenize(sent) for sent in Knicks_History_sentences]
```

```python
#Tag all words in all sentences with POS tags
Monty_Python_sentences_tagged = [nltk.pos_tag(sent) for sent in Monty_Python_sentences]
Knicks_History_sentences_tagged = [nltk.pos_tag(sent) for sent in Knicks_History_sentences]

#Find all sentence indices that contains words with adjective and/or adverb POS tags
Monty_filter_indices = set()
for sent_index in range(len(Monty_Python_sentences_tagged)):
    for (word,tag) in Monty_Python_sentences_tagged[sent_index]:
        if tag in ('JJ','JJR','JJS','RB','RBR','RBS','WRB'): Monty_filter_indices.add(sent_index)
Knicks_filter_indices = set()
for sent_index in range(len(Knicks_History_sentences_tagged)):
    for (word,tag) in Knicks_History_sentences_tagged[sent_index]:
        if tag in ('JJ','JJR','JJS','RB','RBR','RBS','WRB'): Knicks_filter_indices.add(sent_index)


#Filter to only sentences that have adjective and/or adverb POS tags
Monty_filtered = [sent for sent in Monty_Python_sentences_tagged if Monty_Python_sentences_tagged.index(sent) in Monty_filter_indices]
Knicks_filtered = [sent for sent in Knicks_History_sentences_tagged if Knicks_History_sentences_tagged.index(sent) in Knicks_filter_indices]

#Number of sentences in each step
#Monty Python
print('SUMMARY OF INITIAL TOKENIZATION\n\nMonty Python and The Holy Grail\nFull Text:', len(Monty_Python_sentences), 'sentences')
print('Filtered Text:', len(Monty_filtered), 'sentences', len(Monty_Python_sentences) - len(Monty_filtered), 'removed')

#Knicks_History
print('\n\nKnicks History\nFull Text:', len(Knicks_History_sentences), 'sentences')
print('Filtered Text:', len(Knicks_filtered), 'sentences', len(Knicks_History_sentences) - len(Knicks_filtered), 'removed')
```

Metric Set 1: 
<br>
Total number of sentnences, words, characters, adjectve phrases, and adverb phrases
<br>
<br>
We can find these using the filtered sentences and performing calcualtions on them.

```python
#Total number of sentences, words, characters, adjectives, and adverbs
import pandas
DF1 = pandas.DataFrame(index = ['Sentences', 'Words', 'Characters', 'Adjectives', 'Adverbs'], columns = ['Filtered Monty Python','Filtered Knicks History'])
#Total number of sentences
DF1.loc['Sentences']['Filtered Monty Python'] = len(Monty_filtered)
DF1.loc['Sentences']['Filtered Knicks History'] = len(Knicks_filtered)
#Total number of words
DF1.loc['Words']['Filtered Monty Python'] = sum([len(sent) for sentDF1 in Monty_filtered])
DF1.loc['Words']['Filtered Knicks History'] = sum([len(sent) for sentDF1 in Knicks_filtered])
#Total number of characters
DF1.loc['Characters']['Filtered Monty Python'] = sum([len(word) for (word,tag) in sent for sent in Monty_filtered])
DF1.loc['Characters']['Filtered Knicks History'] = sum([len(word) for (word,tag) in sent for sent in Knicks_filtered])
#Total number of adjectives
DF1.loc['Adjectives']['Filtered Monty Python'] = sum([len(word[0]) for (word,tag) in sent for sent in Monty_filtered if tag in ('JJ', 'JJR', 'JJS')])
DF1.loc['Adjectives']['Filtered Knicks History'] = sum([len(word[0]) for (word,tag) in sent for sent in Knicks_filtered if tag in ('JJ', 'JJR', 'JJS')])
#Total number of adverbs
DF1.loc['Adverbs']['Filtered Monty Python'] = sum([len(word[0]) for (word,tag) in sent for sent in Monty_filtered if tag in ('RB', 'RBR', 'RBS', 'WRB')])
DF1.loc['Adverbs']['Filtered Knicks History'] = sum([len(word[0]) for (word,tag) in sent for sent in Knicks_filtered if tag in ('RB', 'RBR', 'RBS', 'WRB')])
#Display results
DF1
```

Metric Set 2: 
<br>
Min, Max, and Average sentence length measured both in words and characters
<br>
<br>
We will once again use the filtered sentences to calcaulate the lenghts in words, but will need to calculate the number of characters per sentence to find the sentence lengths in characters.

```python
#Min, Max, Average sentences length in both words and characters
DF2 = pandas.DataFrame(index = ['Min Sentence Length (Words)', 'Max Sentence Length (Words)', 'Average Sentence Length (Words)',
                                'Min Sentence Length (Characters)', 'Max Sentence Length (Characters)', 'Average Sentence Length (Characters)']
                                , columns = ['Filtered Monty Python','Filtered Knicks History'])
#Sentences lengths (words)
DF2.loc['Min Sentence Length (Words)']['Filtered Monty Python'] = min([len(sent) for sent in Monty_filtered])
DF2.loc['Max Sentence Length (Words)']['Filtered Monty Python'] = max([len(sent) for sent in Monty_filtered])
DF2.loc['Average Sentence Length (Words)']['Filtered Monty Python'] = sum([len(sent) for sent in Monty_filtered]) / len(Monty_filtered)
DF2.loc['Min Sentence Length (Words)']['Filtered Knicks History'] = min([len(sent) for sent in Knicks_filtered])
DF2.loc['Max Sentence Length (Words)']['Filtered Knicks History'] = max([len(sent) for sent in Knicks_filtered])
DF2.loc['Average Sentence Length (Words)']['Filtered Knicks History'] = sum([len(sent) for sent in Knicks_filtered]) / len(Knicks_filtered)
#Sentence lengths (characters)
#Find characters per sentence
Monty_filtered_sent_chars = []
for sent in Monty_filtered: Monty_filtered_sent_chars.append(sum([len(word) for (word,tag) in sent]))
Knicks_filtered_sent_chars = []
for sent in Knicks_filtered: Knicks_filtered_sent_chars.append(sum([len(word) for (word,tag) in sent]))
#Calculate sentence lengths
DF2.loc['Min Sentence Length (Characters)']['Filtered Monty Python'] = min(Monty_filtered_sent_chars)
DF2.loc['Max Sentence Length (Characters)']['Filtered Monty Python'] = max(Monty_filtered_sent_chars)
DF2.loc['Average Sentence Length (Characters)']['Filtered Monty Python'] = sum(Monty_filtered_sent_chars) / len(Monty_filtered_sent_chars)
DF2.loc['Min Sentence Length (Characters)']['Filtered Knicks History'] = min(Knicks_filtered_sent_chars)
DF2.loc['Max Sentence Length (Characters)']['Filtered Knicks History'] = max(Knicks_filtered_sent_chars)
DF2.loc['Average Sentence Length (Characters)']['Filtered Knicks History'] = sum(Knicks_filtered_sent_chars) / len(Knicks_filtered_sent_chars)
#Display results
DF2
```

Metric Set 3: 
<br>
Min, Max, and Average number of adjectives and adverbs per sentence
<br>
<br>
Much like the characters per sentence for the sentence lengths, we will need to find the adjectives and adverbs per sentence for this analysis.

```python
#Min, Max, Average adjectives and adverbs per sentence
DF3 = pandas.DataFrame(index = ['Min Adjectives in Sentence', 'Max Adjectives in Sentence', 'Average Adjectives in Sentence',
                                'Min Adverbs in Sentence', 'Max Adverbs in Sentence', 'Average Adverbs in Sentence']
                                , columns = ['Filtered Monty Python','Filtered Knicks History'])
#Find adjectives and adverbs per sentence
Monty_filtered_sent_adjs = []
for sent in Monty_filtered: Monty_filtered_sent_adjs.append(sum([len(word[0]) for (word,tag) in sent if tag in ('JJ', 'JJR', 'JJS')]))
Knicks_filtered_sent_adjs = []
for sent in Knicks_filtered: Knicks_filtered_sent_adjs.append(sum([len(word[0]) for (word,tag) in sent if tag in ('JJ', 'JJR', 'JJS')]))
Monty_filtered_sent_advs = []
for sent in Monty_filtered: Monty_filtered_sent_advs.append(sum([len(word[0]) for (word,tag) in sent if tag in ('RB', 'RBR', 'RBS', 'WRB')]))
Knicks_filtered_sent_advs = []
for sent in Knicks_filtered: Knicks_filtered_sent_advs.append(sum([len(word[0]) for (word,tag) in sent if tag in ('RB', 'RBR', 'RBS', 'WRB')]))

#Calculate metrics
DF3.loc['Min Adjectives in Sentence']['Filtered Monty Python'] = min(Monty_filtered_sent_adjs)
DF3.loc['Max Adjectives in Sentence']['Filtered Monty Python'] = max(Monty_filtered_sent_adjs)
DF3.loc['Average Adjectives in Sentence']['Filtered Monty Python'] = sum(Monty_filtered_sent_adjs) / len(Monty_filtered_sent_adjs)
DF3.loc['Min Adverbs in Sentence']['Filtered Monty Python'] = min(Monty_filtered_sent_advs)
DF3.loc['Max Adverbs in Sentence']['Filtered Monty Python'] = max(Monty_filtered_sent_advs)
DF3.loc['Average Adverbs in Sentence']['Filtered Monty Python'] = sum(Monty_filtered_sent_advs) / len(Monty_filtered_sent_advs)
DF3.loc['Min Adjectives in Sentence']['Filtered Knicks History'] = min(Knicks_filtered_sent_adjs)
DF3.loc['Max Adjectives in Sentence']['Filtered Knicks History'] = max(Knicks_filtered_sent_adjs)
DF3.loc['Average Adjectives in Sentence']['Filtered Knicks History'] = sum(Knicks_filtered_sent_adjs) / len(Knicks_filtered_sent_adjs)
DF3.loc['Min Adverbs in Sentence']['Filtered Knicks History'] = min(Knicks_filtered_sent_advs)
DF3.loc['Max Adverbs in Sentence']['Filtered Knicks History'] = max(Knicks_filtered_sent_advs)
DF3.loc['Average Adverbs in Sentence']['Filtered Knicks History'] = sum(Knicks_filtered_sent_advs) / len(Knicks_filtered_sent_advs)

#Display results
DF3
```

Metric Set 4: 
<br>
Min, Max, and Average word length for adjectives and adverbs
<br>
<br>
This step will require we first find all adjectives and all adverbs separately.

```python
#Min, Max, Average word lengths adjectives and adverbs
DF4 = pandas.DataFrame(index = ['Min Adjective Length', 'Max Adjective Length', 'Average Adjective Length',
                                'Min Adverb Length', 'Max Adverb Length', 'Average Adverb Length']
                                , columns = ['Filtered Monty Python','Filtered Knicks History'])
#Find adverbs per sentence
#Monty adjectives
Monty_adjs = []
for sent in Monty_filtered: 
    for (word,tag) in sent:
        if tag in ('JJ', 'JJR', 'JJS'): Monty_adjs.append(word)
#Knicks adjectives
Knicks_adjs = []
for sent in Knicks_filtered: 
    for (word,tag) in sent:
        if tag in ('JJ', 'JJR', 'JJS'): Knicks_adjs.append(word)
#Monty adverbs
Monty_advs = []
for sent in Monty_filtered: 
    for (word,tag) in sent:
        if tag in ('RB', 'RBR', 'RBS', 'WRB'): Monty_advs.append(word)
#Knicks adverbs
Knicks_advs = []
for sent in Knicks_filtered: 
    for (word,tag) in sent:
        if tag in ('RB', 'RBR', 'RBS', 'WRB'): Knicks_advs.append(word)

#Calculate metrics
DF4.loc['Min Adjective Length']['Filtered Monty Python'] = min([len(word) for word in Monty_adjs])
DF4.loc['Max Adjective Length']['Filtered Monty Python'] = max([len(word) for word in Monty_adjs])
DF4.loc['Average Adjective Length']['Filtered Monty Python'] = sum([len(word) for word in Monty_adjs]) / len(Monty_adjs)
DF4.loc['Min Adverb Length']['Filtered Monty Python'] = min([len(word) for word in Monty_advs])
DF4.loc['Max Adverb Length']['Filtered Monty Python'] = max([len(word) for word in Monty_advs])
DF4.loc['Average Adverb Length']['Filtered Monty Python'] = sum([len(word) for word in Monty_advs]) / len(Monty_advs)
DF4.loc['Min Adjective Length']['Filtered Knicks History'] = min([len(word) for word in Knicks_adjs])
DF4.loc['Max Adjective Length']['Filtered Knicks History'] = max([len(word) for word in Knicks_adjs])
DF4.loc['Average Adjective Length']['Filtered Knicks History'] = sum([len(word) for word in Knicks_adjs]) / len(Knicks_adjs)
DF4.loc['Min Adverb Length']['Filtered Knicks History'] = min([len(word) for word in Knicks_advs])
DF4.loc['Max Adverb Length']['Filtered Knicks History'] = max([len(word) for word in Knicks_advs])
DF4.loc['Average Adverb Length']['Filtered Knicks History'] = sum([len(word) for word in Knicks_advs]) / len(Knicks_advs)

#Display results
DF4
```
